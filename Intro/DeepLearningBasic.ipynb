#### Different algorithms used in deep learning 
Deep Learning method use neural network architectures, that why Deep Learning models are also referred as deep neural networks. The term deep refers number of hidden layers in neural network. Traditional neural networks include 2-3 hidden layers, while deep networks can have as many as 150 hidden layers. Deep Learning models are trained by using large set of labeled data and neural network architectures that learn features directly from the data without the need for manual features extractions.

#### There four fundamental network architectures:
- Unsupervised Pre-trained Networks
- Convolutional Neural Networks
- Recurrent Neural Networks
- Recursive Neural Networks

#### Working Principles
- Deep learning maps inputs to outputs. It finds correlations. It is known as a “universal approximator”.
- It learn to approximate an unknown function f(x) = y between any input x and any output y.
- In the process of learning, a neural network finds the right f, or the correct manner of transforming x into y.

#### Optimizers: 
##### Optimizers update the weight parameters to minimize the loss function. Loss function acts as guides to the terrain telling optimizer if it is moving in the right direction to reach the bottom of the valley, the global minimum
- Feedforward Networks
- Gradient Descent
- SGD(Stochastic gradient descent)
- AdaGrad (also optimizes learning rate)
- Adadelta
- RMSPROP

#### Activation Function:
##### In a neural network, the activation function is responsible for transforming the summed weighted input from the node into the activation of the node or output for that input.
- ReLU:(rectified linear unit); 
    f(z) = max(0,z); df(z)/dz = {0,1}
    Solves Vanising gradient problem, unless the dereivative comes 0
- RRELU
- SIGMOID (for binary classificaton)
    f(z)=1/(1-e^-y) , df(z)/dz = {0,0.25} 
    so in backward propagation it causes Vanishing Gradient Descent Problem
- SOFTMAX (for multi classification)
- SOFTPLUS
- SOFTSIGN
- TANH
    f(z)=(1-e^-2z)/(1+e^-2z);
- CUBE
- ELU
- HARDSIGMOID
- HARDTANH
- IDENTITY
- LEAKYRELU 
    upgraded version of ReLU to solve ReLU problem
    f(z) = max(0.01,z); df(z)/dz= {0.01,1}
- RATIONALTANH

#### Loss Function: summation(actual -predicted)^2 
##### A loss function is used to optimize the parameter values in a neural network model. 
##### Loss functions map a set of parameter values for the network onto a scalar value that indicates how well those parameter accomplish the task the network is intended to do
- MSE: Mean Squared Error: Linear Regression
- EXPLL: Exponential log likelihood: Poisson Regression
- XENT: Cross Entropy: Binary Classification
- MCXENT: Multiclass Cross Entropy
- RMSE_XENT: RMSE Cross Entropy
- SQUARED_LOSS: Squared Loss
- NEGATIVELOGLIKELIHOOD: Negative Log Likelihood

#### Plcaceholder
Placeholder are used which allows your graph to take external inputs as parameters.
it is used to provide a value later or during runtime

```python
#Creating Placeholder
a=tf.placeholder(tf.float32)
b=tf.placeholder(tf.float32)
#assigning multiplication operation wrt a and b to node mul
mul=a*b
```

```python
#creating sesson object
sess=tf.Session()
```

```python
#Executing mul by passing values for a and b
output=sess.run(mul,{a:[1,3], b:[2,4]})
print('Multiplying a and b :', output)
```

#### Variables
variables allows to ass such parameter or node to the graph that are trainable i.e the value can be modified over the period of a time

```python
#Creating  variables for parameter slope(W) with initial value as 0.4
W=tf.Variable([0.4],tf.float32)
```

```python
#Creating  variables for parameter bias(b) with initial value as -0.4
b=tf.Variable([-0.4],tf.float32)
```

```python
#Creating placeholder for providing input or independent variable, denoted by x
x=tf.placeholder(tf.float32)
```

```python
#equation for Linear regression
lm=W*x + b
```

```python
#Initialising all variables
sess=tf.Session()
init=tf.global_variables_initializer()
sess.run(init)
```

```python
#Running regression model to calculate output wrt provide x values
print(sess.run(lm,{x:[1,2,3,4]}))
```

```python
#Model validation
#calculating loss
y=tf.placeholder(tf.float32)
error=lm-y
sqr_error=tf.square(error)
loss=tf.reduce_sum(sqr_error)
print(sess.run(loss, {x:[1,2,3,4],y:[2,4,6,8]}))
```

```python
#Creating an instance of gradient descent optimizer
optimizer = tf.train.GradientDescentOptimizer(0.01)#Learning rate taken  0.01
train=optimizer.minimize(loss)
```

```python
for i in range(1000):
    sess.run(train,{x:[1,2,3,4],y:[2,4,6,8]})
print(sess.run([W,b]))
```
